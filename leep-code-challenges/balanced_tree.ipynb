Given a binary tree, determine if it is  height-balanced

Example 1:


* Input: root = [3,9,20,null,null,15,7]
* Output: true

Example 2:


* Input: root = [1,2,2,3,3,null,null,4,4]
* Output: false

Example 3:

* Input: root = []
* Output: true
 

Constraints:

The number of nodes in the tree is in the range [0, 5000].
-104 <= Node.val <= 104

```python
from typing import Optional

#Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution:
    def isBalanced(self, root: Optional[TreeNode]) -> bool:
        # handle the root null exception case
        if( root is None ):
            return True
        
        def node_depth(root):
            if( root is None ):
                return -1 , True
            
            # check on root node left & right branches
            left_depth , left_balance = node_depth(root.left)
            right_depth , right_balance = node_depth(root.right)
            if( not( right_balance and  left_balance  ) ):
                return 0 , False
            if( abs( left_depth - right_depth ) <= 1 ):
                return max( [left_depth , right_depth] ) + 1 , True
            else:
                return max( [left_depth , right_depth] ) + 1 , False
        max_depth , balance_state  = node_depth(root)
        return balance_state







                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   
```
