You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

Example 1:




* Input: l1 = [2,4,3], l2 = [5,6,4]
* Output: [7,0,8]

Explanation: 342 + 465 = 807.

Example 2:

* Input: l1 = [0], l2 = [0]
* Output: [0]

Example 3:

* Input: l1 = [9,9,9,9,9,9,9], l2 = [9,9,9,9]
* Output: [8,9,9,9,0,0,0,1]
 

Constraints:

1) The number of nodes in each linked list is in the range [1, 100].
2) 0 <= Node.val <= 9
4) It is guaranteed that the list represents a number that does not have leading zeros.

```python
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        # l1 / l2 are the linklists output also should be linklist
        # l1 / l2 can be in different sizes 
        # linklist element traversel O(n)
        # addition perform in reverse order ( otherwise have to reverse every thing )

        # define the head node
        head , carry = ListNode(0) , 0
        head_ref= head
        while ( ( l1 is not None ) or ( l2 is not None ) or ( carry != 0 ) ):

            # check none condition for each
            l1_val = l1.val if l1 is not None else 0
            l2_val = l2.val if l2 is not None else 0

            l1 = l1.next if l1 is not None else None 
            l2 = l2.next if l2 is not None else None

            sum = l1_val + l2_val + carry
            carry , remainder = sum % 10 , sum // 10 

            # define the new node
            temp_node = ListNode( remainder )
            head_ref.next = temp_node
            head_ref = temp_node

        return head.next
```
