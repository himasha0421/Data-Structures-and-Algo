Given the root of a binary tree, invert the tree, and return its root.

 

Example 1:



* Input: root = [4,2,7,1,3,6,9]
* Output: [4,7,2,9,6,3,1]


Example 2:


* Input: root = [2,1,3]
* Output: [2,3,1]

Example 3:

* Input: root = []
* Output: []
 

Constraints:

The number of nodes in the tree is in the range [0, 100].

-100 <= Node.val <= 100

```python
from typing import Optional

#Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        """
        check initial condition on root node null
        initate traversal on depth first search using stack
        swap children after comming to a node
        """
        if(root is None): return root

        stack =[root]
        while len(stack) != 0 :
            selected_node = stack.pop()
            # swap children
            temp_node = selected_node.left
            selected_node.left = selected_node.right
            selected_node.right = temp_node

            # check on the children state
            if( selected_node.left is not None):
                stack.append( selected_node.left )
            if( selected_node.right is not None ):
                stack.append( selected_node.right )

        return root

```

```python
# recurssive method

class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        """
        check initial condition on root node null
        initate traversal on depth first search using stack
        swap children after comming to a node
        """
        if(root is None): return root

        temp_node = root.left
        # swap children
        root.left = root.right
        root.right = temp_node

        # then recursive over next levl
        self.invertTree(root.left)
        self.invertTree(root.right)

        return root
```
