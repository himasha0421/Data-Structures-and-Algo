```
Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.

Note that you must do this in-place without making a copy of the array.

 

Example 1:

Input: nums = [0,1,0,3,12]
Output: [1,3,12,0,0]
Example 2:

Input: nums = [0]
Output: [0]
 

Constraints:

1 <= nums.length <= 104
-231 <= nums[i] <= 231 - 1
 

Follow up: Could you minimize the total number of operations done?
```

````python
class Solution(object):
    def moveZeroes(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """

        """
        no extra array/stack/queue
        two pinters to track start/end 0/1
        if left == 0 & right !=0 :
            then swap(left,right)
            start -> +1 / end -> +1
        if left ==0 & right ==0 :
            end -> +1
        if left !=0 & right ==0 :
            start -> +1 / end -> +1

        """
        # handle len(nums) <= 1: return nums no swap
        if( len(nums) <= 1 ): return nums

        start , end = 0 , 1

        while len(nums) != end :
            if( ( nums[start] == 0 ) and ( nums[end] ) != 0  ): 
                temp_val = nums[end]
                nums[end] = nums[start]
                nums[start] = temp_val
                start +=1 
                end += 1

            elif( ( nums[start] == 0 ) and ( nums[end]==0 ) ):
                end += 1

            else:
                start += 1
                end += 1

        return nums
        
````

````python
class Solution(object):
    def moveZeroes(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        
        """
        keep track of zeros size previously handled (hit)
        if current_val !=0 and hit>0 : do a element swap
        """

        hit =0 

        for i in range(len(nums)):
            if( nums[i] ==0 ): hit +=1

            elif( (nums[i] != 0 ) and (hit >0) ):
                temp_val = nums[i]
                nums[i] = nums[i-hit]
                nums[i-hit] = temp_val
        return nums
        
````
